## Bonus task: Image dataset from directory

### Working with keras images dataset from directory

```python
#import needed libs
import pathlib
import matplotlib.pyplot as plt
import numpy as np
import os
import imageio
import PIL
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
#working iwth keras dataset creation of 80/20
datadir = "C:\\Users\\camil\\Desktop\\my own study\\studying the cnn\\final"
#i decided to create some dataset from image folders we have, and read some images
data_dir = pathlib.Path(datadir)
```

```python
#creation of dataset 80/20
training = tf.keras.preprocessing.image_dataset_from_directory(data_dir, validation_split=0.2, subset="training", seed=123,
  image_size=(180, 180), batch_size=32)
#create validation for it
validation = tf.keras.preprocessing.image_dataset_from_directory(data_dir, validation_split=0.2, subset="validation", seed=123,
  image_size=(180, 180), batch_size=32)
```

```python
#working iwth keras dataset creation of 70/30
training70 = tf.keras.preprocessing.image_dataset_from_directory(data_dir, validation_split=0.3, subset="training", seed=123,
  image_size=(180, 180), batch_size=32)
#create validation
validation70 = tf.keras.preprocessing.image_dataset_from_directory(data_dir, validation_split=0.3, subset="validation", 
                                                                   seed=123, image_size=(180, 180), batch_size=32)
```

```python
#add normilized layer
normalization_layer = layers.experimental.preprocessing.Rescaling(1./255)
```

```python
#dataset 80/20
normalized = training.map(lambda x, y: (normalization_layer(x), y))
image_batch, labels_batch = next(iter(normalized))
first_image = image_batch[0]
# Notice the pixels values are now in `[0,1]`.
print(np.min(first_image), np.max(first_image))
```

```python
#dataset 70/30
normalized_70 = training70.map(lambda x, y: (normalization_layer(x), y))
image_batch70, labels_batch70 = next(iter(normalized_70))
first_image70 = image_batch70[0]
# Notice the pixels values are now in `[0,1]`.
print(np.min(first_image70), np.max(first_image70))
```

```python
#add regularization
data_augmentation = keras.Sequential(
  [
    layers.experimental.preprocessing.RandomFlip("horizontal", input_shape=(180, 180, 3)),
    layers.experimental.preprocessing.RandomRotation(0.1),
    layers.experimental.preprocessing.RandomZoom(0.1),
  ]
)
```

```python
#show some results of 80/20 dataset
plt.figure(figsize=(10, 10))
for images, _ in training.take(1):
    for i in range(9):
        augmented_images = data_augmentation(images)
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(augmented_images[0].numpy().astype("uint8"))
        plt.axis("off")
```

```python
#explore some results of 70/30 dataset
plt.figure(figsize=(10, 10))
for images, _ in training70.take(1):
    for i in range(9):
        augmented_images = data_augmentation(images)
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(augmented_images[0].numpy().astype("uint8"))
        plt.axis("off")
```

### Model defining and fitting on custom model

```python
#define easy model from scratch 
model = Sequential([
  data_augmentation,
  layers.experimental.preprocessing.Rescaling(1./255),
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Dropout(0.2),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(5)
])
#compile model
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
#see what we have
model.summary()
```

```python
#history fitting on 80/20 dataset
history = model.fit(training, validation_data=validation,epochs=5)
```

```python
#graph plotting on 20/80 dataset
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(5)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
#model fitting on 70/30 dataset
history70 = model.fit(training70, validation_data=validation70,epochs=5)
```

```python
#graph plotting on 70/30 dataset
acc = history70.history['accuracy']
val_acc = history70.history['val_accuracy']

loss = history70.history['loss']
val_loss = history70.history['val_loss']

epochs_range = range(5)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
