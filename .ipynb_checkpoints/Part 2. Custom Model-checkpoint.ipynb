### Part 2. Working with Custom model

```python
#import all libs i needed
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
import tensorflow as tf
import keras
import os
import time
from PIL import Image 
import matplotlib.pyplot as plt
import numpy as np
import cv2
from keras.utils import to_categorical
import seaborn as sn
from sklearn.model_selection import train_test_split
from keras.optimizers import SGD, Adam
from keras.regularizers import l2
from keras.models import load_model, Sequential
import keras_preprocessing
from sklearn.utils import shuffle
from keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout, MaxPooling2D, BatchNormalization
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
#clean all umimportant data for memory usage
import gc
```

```python
#define my values and path before working with it
x_data = []
y_data = []
datadir = "C:\\Users\\camil\\Desktop\\my own study\\studying the cnn\\final"
categories = ["accordeon", "dombyra", "kobyz"]
```

```python
def create_dataset(cats, img_folder):#this function creates dataset for me, by reading all the data in 3 folders
    img_data_array=[]
    class_name=[]
    c = 0
    l = dict()
    r = dict()

    for j in cats:#save the temp list to work with
        if not j.startswith('.'): 
            l[j] = c
            r[c] = j
            c +=1
    
    for j in categories:
        count = 0
        path = os.path.join(datadir, j)
        for k in os.listdir(path):
            isExist =os.path.exists(path+"\\"+k)#like 10th or 12th images 
            if isExist==False:#check if image exist
                continue#and here i get all my images
            else:
                img = Image.open(path + "\\" + k).convert('L')#grayscale
                img = img.resize((120, 224))#get  better shape
                arr = np.array(img)#save as array
                img_data_array.append(arr)# so the we could use it in list
                count = count + 1
        values = np.full((count, 1), l[j]) 
        class_name.append(values)#get my labels data
    #get my data corrected: normilize and reshaped
    img_data_array = np.array(img_data_array, dtype = 'float32')
    class_name = np.array(class_name)   
    class_name = class_name.reshape(3, 1)   
    return img_data_array, class_name

#call create_dataset function
x_data, y_data = create_dataset(categories, datadir)
```

```python
#check the shape of each dataset
print("Y dataset shape: ", y_data.shape)
print("X dataset shape: ", x_data.shape)
```

```python
#convert our label data
y = to_categorical(y_data.shape, 1660)
print("Y converted dataset shape: ", y.shape)
```

```python
#reshape Y dataset so we could work with it
Y =y.reshape((332, 10))
print("Y converted dataset reshaped: ", Y.shape)
```

```python
#reshape our data
x_data = x_data.reshape((332, 120, 224, 1))
#normalize dataset
x_data = x_data / 255.0
#see the result
print("X normalized-dataset shape: ", x_data.shape)
```

```python
#we should randomly divide data: we could shuffle before or use in train_test_split
x_data, Y = shuffle(x_data, Y)
```

```python
#split to train and test dataset on 70/30
x_train,x_test,y_train,y_test=train_test_split(x_data,Y, train_size=0.7, test_size=0.3, shuffle=True)
#define validation data
x_validate,x_val_test,y_validate,y_val_test = train_test_split(x_test, y_test, #stratify=y_test, #if needed
                                                               test_size = 0.55, random_state=42) 
```

```python
#lets look with what we are working of 70/30 data
print('Y-Train dataset shape: ', y_train.shape)
print('Y-Test dataset shape: ', y_test.shape)
print('X-Train dataset shape: ', x_train.shape)
print('X-Test dataset shape: ', x_test.shape)
#lets print validation data quantity
print('Y-Validate test dataset samples: ', y_validate.shape[0])
print('X-Validate train dataset samples: ', x_validate.shape[0])
```

```python
#split to train and test dataset on 80/20
x_80train,x_20test,y_80train,y_20test=train_test_split(x_data, Y,train_size=0.8, test_size=0.2, shuffle=True)
#define validation data
x_80validate,x_val_20test,y_80validate,y_val_20test = train_test_split(x_20test,y_20test, #stratify=y_20test, #if needed
                                                                       test_size = 0.45,  random_state=42)
```

```python
#lets look with what we are working with 80/20 data
print('Y-Train dataset shape: ', y_80train.shape)
print('Y-Test dataset shape: ', y_20test.shape)
print('X-Train dataset shape: ', x_80train.shape)
print('X-Test dataset shape: ', x_20test.shape)
#print validation data samples value
print('Y-Validate test dataset samples: ', y_80validate.shape[0])
print('X-Validate train dataset samples: ', x_80validate.shape[0])
```

```python
#clear unecessary data after using it
gc.collect()
del y_data
```

```python
# build new model - model
#reg=l2(1e-4)   # L2 or "ridge" regularisation
reg2=None  #with no regularization
num_filters2=32
ac2='relu'
adm2=Adam(lr=0.001,decay=0, beta_1=0.9, beta_2=0.999, epsilon=1e-08)
opt2=adm2
drop_dense2=0.5
drop_conv2=0

model = Sequential()

model.add(Conv2D(num_filters2, (3, 3), activation=ac2, kernel_regularizer=reg2, input_shape=(120, 224, 1),
                  padding='same'))
model.add(BatchNormalization(axis=-1))
model.add(Conv2D(num_filters2, (3, 3), activation=ac2,kernel_regularizer=reg2,padding='same'))
model.add(BatchNormalization(axis=-1))
model.add(MaxPooling2D(pool_size=(2, 2)))   # reduces to 16x16x3xnum_filters
model.add(Dropout(drop_conv2))

model.add(Conv2D(2*num_filters2, (3, 3), activation=ac2,kernel_regularizer=reg2,padding='same'))
model.add(BatchNormalization(axis=-1))
model.add(Conv2D(2*num_filters2, (3, 3), activation=ac2,kernel_regularizer=reg2,padding='same'))
model.add(BatchNormalization(axis=-1))
model.add(MaxPooling2D(pool_size=(2, 2)))   # reduces to 8x8x3x(2*num_filters)
model.add(Dropout(drop_conv2))

model.add(Conv2D(4*num_filters2, (3, 3), activation=ac2,kernel_regularizer=reg2,padding='same'))
model.add(BatchNormalization(axis=-1))
model.add(Conv2D(4*num_filters2, (3, 3), activation=ac2,kernel_regularizer=reg2,padding='same'))
model.add(BatchNormalization(axis=-1))
model.add(MaxPooling2D(pool_size=(2, 2)))   # reduces to 4x4x3x(4*num_filters)
model.add(Dropout(drop_conv2))

model.add(Flatten())
model.add(Dense(512, activation=ac2,kernel_regularizer=reg2))
model.add(BatchNormalization())
model.add(Dropout(drop_dense2))
model.add(Dense(10, activation='softmax'))
#compiling model
model.compile(optimizer=opt2, loss='categorical_crossentropy', metrics=['accuracy'])
#showing sum of model
model.summary()
```

```python
#data augmenatation
datagen = ImageDataGenerator(
        featurewise_center=False,  # set input mean to 0 over the dataset
        samplewise_center=False,  # set each sample mean to 0
        featurewise_std_normalization=False,  # divide inputs by std of the dataset
        samplewise_std_normalization=False,  # divide each input by its std
        zca_whitening=False,  # apply ZCA whitening
        rotation_range = 30,  # randomly rotate images in the range (degrees, 0 to 180)
        zoom_range = 0.2, # Randomly zoom image 
        width_shift_range=0.1,  # randomly shift images horizontally (fraction of total width)
        height_shift_range=0.1,  # randomly shift images vertically (fraction of total height)
        horizontal_flip = True,  # randomly flip images
        vertical_flip=False)  # randomly flip images
```

```python
#add regularization
train_70generator = datagen.flow(x=x_train, y=y_train)
train_80generator = datagen.flow(x=x_80train, y=y_80train)
```

```python
#create location of my log for tensorboard use
root_logdir = os.path.join(os.curdir, "logs\\fit\\")

#fuction to get it
def get_run_logdir():
    run_id = time.strftime("run_%Y_%m_%d-%H_%M_%S")
    return os.path.join(root_logdir, run_id)
#after it, i run the function to get my  log
run_logdir = get_run_logdir()
#where i call it and save as callback for later use
tensorboard_cb = keras.callbacks.TensorBoard(run_logdir)
```

```python
#fit our model on new dataset 70/30
history_model1 = model.fit(train_70generator, batch_size=20,
                           epochs=5, validation_data=(x_validate, y_validate),
                     verbose = 1, callbacks=[tensorboard_cb]) 
```

```python
#see the outcome of 70/30
acc = history_model1.history['accuracy']
val_acc = history_model1.history['val_accuracy']
loss = history_model1.history['loss']
val_loss = history_model1.history['val_loss']

epochs_range = range(5)

plt.figure(figsize=(15, 15))
plt.subplot(2, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(2, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
#fit our model to 80/20
history_model2 = model.fit(train_80generator, batch_size = 16, epochs=5, validation_data=(x_80validate, y_80validate),
                     verbose = 1, callbacks=[tensorboard_cb]) 
```

```python
#graph result of 80/20
acc = history_model2.history['accuracy']
val_acc = history_model2.history['val_accuracy']
loss = history_model2.history['loss']
val_loss = history_model2.history['val_loss']

epochs_range = range(5)

plt.figure(figsize=(15, 15))
plt.subplot(2, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(2, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
#calculate loss and accuracy on 70/30 test data
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print('Test accuracy: {:2.2f}%'.format(test_accuracy*100))
#calculate loss and accuracy on 80/20 test data
test_loss, test_accuracy = model.evaluate(x_20test, y_20test)
print('Test accuracy: {:2.2f}%'.format(test_accuracy*100))
#why so lost?
#needed to do fine-tuning the top layers will directly lead to overfitting. 
#Load imagenet weights but re-train all or couple of layers.
#also could try label smoothing. It can also help your classifier to give more probability to the correct class
```

```python
#i wanted to plot confusion matrix but it could not becasue did not accept multiclass and binary
#so stackoverlow says: that by predicting my classes separately and rounding them will help
#because both of them y_test and x_test are one-hot encoded vectors
rounded_predictions_70 = model.predict_classes(x_test, batch_size=128, verbose=0)
rounded_predictions_80 = model.predict_classes(x_20test, batch_size=128, verbose=0)
#see what i got - should be single digit
rounded_predictions_70[1]
rounded_predictions_80[1]
#labes one-hot encoded - need to convert to single digit ones
rounded_labels_70=np.argmax(y_test, axis=1)
rounded_labels_80=np.argmax(y_20test, axis=1)
#checking the result in advance
rounded_labels_70[1]
rounded_labels_80[1]
```

```python
#finally calculating conf matrix of 70/30
confusion_70 = confusion_matrix(rounded_labels_70, rounded_predictions_70, labels=list(np.arange(0,10)))
#conf matrix of 80/20
confusion_80 = confusion_matrix(rounded_labels_80, rounded_predictions_80, labels=list(np.arange(0,10)))
```

```python
#plotting confusion matrix of 70/30
fig, ax = plt.subplots(figsize=(10,8)) 
sn.heatmap(confusion_70, annot=True,ax=ax,fmt="d",linewidths=.5,cmap=sn.color_palette("Blues"));
plt.ylim([ 10,-0.5,]);
plt.ylabel('True label')
plt.xlabel('Predicted label')
```

```python
#plotting confusion matrix of 80/20
fig, ax = plt.subplots(figsize=(10,8)) 
sn.heatmap(confusion_80, annot=True,ax=ax,fmt="d",linewidths=.5,cmap=sn.color_palette("Blues"));
plt.ylim([ 10,-0.5,]);
plt.ylabel('True label')
plt.xlabel('Predicted label')
```

So, here I read data from 3 folders and divided into train and test subsets, where then I created validation dataset to work with, as well add data augmentation and fitted on created from scratch model, which were isnpired by VGG16 model, yet it is not the same one, also, evaluated the result of 70/30 and 80/20 datasets with graphs and metric values.<br>

The reason why my metric values were not nice. Parameter place is huge, while creating from scratch model, so it is important to play with the hyperparameters especially the learning rate, the empirical result shows starting with as low as 1e-6 even helps to converge. Also if you could use some different weight initialization for the weights, that would show tremendous result in terms of convergence, cause the default weight initialization doesn't work well in this case. Lastly, let the model train for longer epochs (like 100) as the space (of parameters) is quite bumpy, you'll see it oscillating a bit but with proper lr it'll converge but take some time.<br>

Learning outcomes:<br>

- use SGD optimiser and set learning rate low
