### Part 3. Working with keras model
##### Note: this is the same work as before, yet I did change the shapes and other values. The reason I did so is it was too much job to rewrite the dataset and change inner values again, so I just created new  dataset to work with from start

```python
#import all libs i needed again
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
import tensorflow as tf
import keras
import os
import time
from PIL import Image 
import matplotlib.pyplot as plt
from matplotlib import image
import matplotlib.image as mpimg 
import numpy as np
import cv2
from keras.utils import to_categorical
import seaborn as sn
import random
from sklearn.model_selection import train_test_split
import h5py
from keras.optimizers import SGD, Adam
from keras.regularizers import l2
from keras.models import load_model, Sequential
import keras_preprocessing
from sklearn.utils import shuffle
from keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout, MaxPooling2D, BatchNormalization
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
from tensorflow.keras import optimizers
#clean all umimportant data for memory usage
import gc
```

```python
#define my values before working with it
x_data = []
y_data = []
datadir = "C:\\Users\\camil\\Desktop\\my own study\\studying the cnn\\final"
categories = ["accordeon", "dombyra", "kobyz"]

#copy the method from previous work here
def create_dataset(cats, img_folder):
    img_data_array=[]
    class_name=[]
    c = 0
    l = dict()
    r = dict()

    for j in cats:
        if not j.startswith('.'): 
            l[j] = c
            r[c] = j
            c +=1
    
    for j in categories:
        count = 0
        path = os.path.join(datadir, j)
        for k in os.listdir(path):
            isExist =os.path.exists(path+"\\"+k)#10,12 
            if isExist==False:
                continue#and here i get all my images
            else:
                img = Image.open(path + "\\" + k).convert('L')
                img = img.resize((132, 320))
                arr = np.array(img)
                img_data_array.append(arr) 
                count = count + 1
        values = np.full((count, 1), l[j]) 
        class_name.append(values)
    #get my data corrected
    img_data_array = np.array(img_data_array, dtype = 'float32')
    class_name = np.array(class_name)   
    class_name = class_name.reshape(3, 1)   
    return img_data_array, class_name

#call for creation dataset
x_data, y_data = create_dataset(categories, datadir)
```

```python
#convert our data
y = to_categorical(y_data.shape, 498)
print("Y converted dataset shape: ", y.shape)
#reshape so we could work with it
Y =y.reshape(332, 3)
print("Y converted dataset reshaped: ", Y.shape)
#reshape our data
x_data = x_data.reshape((332, 110, 128, 3))
#normalize dataset
x_data = x_data / 255.0
#see the result
print("X normalized-dataset shape: ", x_data.shape)
```

```python
#split to train and test dataset
x_train,x_test,y_train,y_test=train_test_split(x_data,Y, train_size=0.7, test_size=0.3, shuffle=True)
#define validation data
x_validate,x_val_test,y_validate,y_val_test = train_test_split(x_test, y_test, #stratify=y_test,
                                                               test_size=0.33,random_state=42) 
```

```python
#lets look with what we are working
print('Y-Train dataset shape: ', y_train.shape)
print('Y-Test dataset shape: ', y_test.shape)
print('X-Train dataset shape: ', x_train.shape)
print('X-Test dataset shape: ', x_test.shape)
#lets print validation data
print('Y-Validate test dataset samples: ', y_validate.shape[0])
print('X-Validate train dataset samples: ', x_validate.shape[0])
```

```python
#split to train and test dataset
x_80train,x_20test,y_80train,y_20test=train_test_split(x_data, Y,train_size=0.8, test_size=0.2, shuffle=True)
#define validation data
x_80validate,x_val_20test,y_80validate,y_val_20test = train_test_split(x_20test,y_20test, #stratify=y_20test, 
                                                                         test_size=0.27, random_state=42)
```

```python
#lets look with what we are working
print('Y-Train dataset shape: ', y_80train.shape)
print('Y-Test dataset shape: ', y_20test.shape)
print('X-Train dataset shape: ', x_80train.shape)
print('X-Test dataset shape: ', x_20test.shape)
#lets print validation data
print('Y-Validate test dataset samples: ', y_80validate.shape[0])
print('X-Validate train dataset samples: ', x_80validate.shape[0])
```

```python
#clear data after using it
del Y
gc.collect()
```

```python
#ResNet152 with keras pre trained weights
resnet152 = tf.keras.applications.ResNet152(include_top=True, weights=None, input_tensor=None,
    input_shape=(110, 128, 3),pooling=None, classes=3,)
#see what we have
resnet152.summary()
#compile model before using it
resnet152.compile(loss='categorical_crossentropy',optimizer = 'adam',metrics=['accuracy'])
```

```python
#add data augmenatation
datagen = ImageDataGenerator(
        featurewise_center=False,  # set input mean to 0 over the dataset
        samplewise_center=False,  # set each sample mean to 0
        featurewise_std_normalization=False,  # divide inputs by std of the dataset
        samplewise_std_normalization=False,  # divide each input by its std
        zca_whitening=False,  # apply ZCA whitening
        rotation_range = 30,  # randomly rotate images in the range (degrees, 0 to 180)
        zoom_range = 0.2, # Randomly zoom image 
        width_shift_range=0.1,  # randomly shift images horizontally (fraction of total width)
        height_shift_range=0.1,  # randomly shift images vertically (fraction of total height)
        horizontal_flip = True,  # randomly flip images
        vertical_flip=False)  # randomly flip images
```

```python
#add regularization
train_70generator = datagen.flow(x=x_train, y=y_train)
train_80generator = datagen.flow(x=x_80train, y=y_80train)
```

```python
#create location of my log for tensorboard use
root_logdir = os.path.join(os.curdir, "logs\\fit\\")

#fuction to get it
def get_run_logdir():
    run_id = time.strftime("run_%Y_%m_%d-%H_%M_%S")
    return os.path.join(root_logdir, run_id)
#after it, i run the function to get my  log
run_logdir = get_run_logdir()
#where i call it and save as callback for later use
tensorboard_cb = keras.callbacks.TensorBoard(run_logdir)
```

```python
#fit our model on new data 70/30
history1 = resnet152.fit(train_70generator, batch_size = 30, epochs=5, validation_data=(x_validate, y_validate),
                     verbose = 1, callbacks=[tensorboard_cb]) 
```

```python
##### see the outcome of 70/30
acc = history1.history['accuracy']
val_acc = history1.history['val_accuracy']
loss = history1.history['loss']
val_loss = history1.history['val_loss']

epochs_range = range(5)

plt.figure(figsize=(15, 15))
plt.subplot(2, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(2, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
#fit our model to new data 80/20 20 18 16
history2 = resnet152.fit(train_80generator, batch_size = 24, epochs=5, validation_data=(x_80validate, y_80validate),
                     verbose = 1, callbacks=[tensorboard_cb]) 
```

```python
#see the outcome of 80/20
acc = history2.history['accuracy']
val_acc = history2.history['val_accuracy']
loss = history2.history['loss']
val_loss = history2.history['val_loss']

epochs_range = range(5)

plt.figure(figsize=(15, 15))
plt.subplot(2, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(2, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
#calculate loss and accuracy on 70/30 test data
test_loss, test_accuracy = resnet152.evaluate(x_validate, y_validate)
print('Test accuracy: {:2.2f}%'.format(test_accuracy*100))
```

```python
#calculate loss and accuracy on 80/20 test data
test_loss, test_accuracy = resnet152.evaluate(x_80validate, y_80validate)
print('Test accuracy: {:2.2f}%'.format(test_accuracy*100))
```

```python
`#one-hot encoded predicted values
rounded_predictions_70 = resnet152.predict(x_test)
rounded_predictions_80 = resnet152.predict(x_20test)
#see what i got - should be single digit
print("70/30: ", rounded_predictions_70[0])
print("80/20: ", rounded_predictions_80[0])
#labes one-hot encoded - need to convert to single digit ones
rounded_labels_70=np.argmax(y_validate, axis =1)
rounded_labels_80=np.argmax(y_20test)
#checking the result in advance
print("70/30: ", rounded_labels_70[0])
print("80/20: ", rounded_labels_80)
```

```python
#finally calculating conf. matrix for 70/30 and 80/20
confusion_70 = confusion_matrix(y_test[0], np.round(rounded_predictions_70[0]), labels=list(np.arange(0,10)))
confusion_80 = confusion_matrix(y_20test[0], np.round(rounded_predictions_80[0]), labels=list(np.arange(0,10)))
```

```python
#lets find something different here
#lets find MAPE, because good rounded labels results
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
#i used precision recall function, now lets fit classification report, because i had such a good results of accuracy
from sklearn.metrics import classification_report
#now lets see result of 70/30
print(classification_report(y_test[0], np.round(rounded_predictions_70[0])))
print("Mape: ", mean_absolute_percentage_error(rounded_labels_70[0], rounded_predictions_70[0]))
```

```python
#values for 80/20 dataset
print(classification_report(y_20test[0], np.round(rounded_predictions_80[0])))
print("Mape: ", mean_absolute_percentage_error(rounded_labels_80, rounded_predictions_80[0]))
#explanation why the results better and why i chose resnet152 are in sum-up section
```
